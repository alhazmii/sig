```python
import warnings
warnings.filterwarnings('ignore')
#memasukan warnings untuk menonaktifkan peringatan yang tidak penting
```

```python
import geopandas as gpd
#memasukan data geopandas alias gpd
```

```python
administrasi_bandung = "https://raw.githubusercontent.com/alhazmii/sig/main/RBI25K_ADMINISTRASI_AR.json"
#menyertakan data, data berupa file .json yang telah diupload di github.com
```

```python
gdf = gpd.read_file(administrasi_bandung)
gdf.head()
#membaca data file yang telah diupload di github.com
#gdf.head digunakan untuk membaca lima baris teratas dari data file
```

```python
gdf = gdf[[ 'NAMOBJ', 'FCODE', 'METADATA', 'WADMKK', 'WADMPR', 'SHAPE_AREA', 'SHAPE_LEN', 'geometry']].fillna(0)
#menyeleksi data yang akan ditampilkan pada folium
```

```python
import folium
m = gdf.explore(column="NAMOBJ", name="Nama Wilayah" )
folium.LayerControl().add_to(m)
m
#memasukan folium
#gdf.explore digunakan untuk menyeleksi data yang akan ditampilkan pada peta
```

```python
m = folium.Map(location = [-6.903272,107.5607547], tiles='cartodb positron', zoom_start=11)
#menampilkan peta daerah Bandung menggunakan titik koordinat, menggunakan jenis peta cartodb positron

tooltip = folium.features.GeoJsonTooltip(fields=['NAMOBJ', 'SHAPE_AREA'],
                                         aliases=["Nama Wlayah", "Daerah Wilayah"],
                                         labels=True,
                                         stick=False)
#menyeleksi data yang akan ditampilkan pada peta
folium.GeoJson(gdf, tooltip=tooltip).add_to(m)
folium.LayerControl().add_to(m)
m
#menampilkan peta
```
